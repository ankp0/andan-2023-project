```python
import numpy as np
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## 1. Предобработка данных

__Имеем табличку с данными о ДТП в Москве (а иногда и в области) за 2022 год. Вид данных оставляет желать лучшего...__

```python
df = pd.read_csv('accidents.csv', sep=';')
df
```

__Дропаем бесполезные колонки__

```python
df = df.drop(['id', 'Широта', 'Долгота', 'ТС свод', 'road_loc', 'pl_status_code', 'Ранено детей до 16', 'Погибло детей до 16'],
             axis=1)
```

__Округ и район - разные показатели. Делим!__

```python
new_df = df['Округ; Район'].str.split(';',expand=True)
new_df = new_df.drop([2, 3], axis=1)
new_df.columns=['Округ','Район']
df.insert(9, 'Округ', new_df['Округ'])
df.insert(10, 'Район', new_df['Район'])
df = df.drop('Округ; Район', axis=1)
```

__Троицкий АО, Новомосковский АО, Троицк и Щербинка входят в один административный округ - ТиНАО. Надо их склеить__

```python
df['Округ'][df['Округ'].isin(['Троицкий АО', 'Новомосковский АО', 'Троицк', 'Щербинка']) == True] = 'Троицкий и Новомосковский АО'
```

__Взявшиеся из ниоткуда квадратные скобочки выглядят некрасиво__

```python
df['Факторы'] = df['Факторы'].str.replace('[','')
df['Факторы'] = df['Факторы'].str.replace(']','')
```

__Часто отсутствуют названия районов/улиц/магистралей, а также данные о массовых ДТП, так как их, логично, случается не так уж и много (особенно хорошо видно на хитмэпе). Посему дропаем колонки с номером транспортного средства больше 3, массовые ДТП рассмотрим как нибудь в другой раз__

```python
df.isnull().sum()
```

```python
sns_heatmap = sns.heatmap(df.sample(5000).isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df = df.drop(['ТС 4', 'ТС 5', 'ТС 6', 'ТС 7', 'ТС 8', 'ТС 9'], axis=1)
```

__Видим, что когда в районе пропуск, это будто быпочти всегда участки обслуживания на территории МО. Ага, так и есть, только одно наблюдение с None в районе лишнее. Игнорируем его и живем счастливо__

```python
df[df['Район'].isnull() == True]
```

```python
df[df['Округ'] == 'Участки обслуживания на территории Московской области']
```

__Иногда название улицы и магистрали идут вместе, хотя понятно, что оно имело место либо по адресу, либо на магистрали. В некоторых случаях точное место ДТП определить можно (например, если это была эстакада над улицей или перекрёсток), но если на месте ДТП не указано никаких объектов, то однозначно понять нельзя (если не смотреть координаты места). Таких наблюдений всего 157 из 7 628, так что удалим их, чтобы не путаться__

```python
df[(df['Назвние магистрали'].isnull() == False) & (df['Улица'].isnull() == False)][['Назвние магистрали', 'Улица', 'Номер Дома',
                                                                                    'Объект УДС на месте ДТП']]
```

## 2. EDA и закономерности

_В данных котегориальных признаков значительно больше числовых, начнём с них_

__Округа от лидеров к лузерам по количеству ДТП:__

```python
df['Округ'].value_counts()
```

__Гисторгамма 10 округов с наибольшис количеством ДТП:__

```python
sns.set(rc={'figure.figsize':(28,10)})
freq = df['Округ'].value_counts()[:10].index.values
sns.histplot(df['Округ'][df['Округ'].isin(freq) == True])
```

__Лидеры по количетсву ДТП среди районов. Забавно, что топ районов по ДТП соверешенно необязательно находятся в округах из топа по ДТП:__

```python
df[['Округ','Район']].value_counts()[:10]
```

__А вот у районов и округов-лузеров корреляция налицо__

```python
df[['Округ','Район']].value_counts()[135:]
```

```python
sns.set(rc={'figure.figsize':(22,10)})
freq1 = df['Район'].value_counts()[:10].index.values
sns.histplot(df['Район'][df['Район'].isin(freq1) == True])
```

```python
sns.set(rc={'figure.figsize':(22,10)})
freq2 = df['Район'].value_counts()[135:].index.values
sns.histplot(df['Район'][df['Район'].isin(freq2) == True])
```

__Топ-10 значений времени, в которое происходили ДТП. Нетрудно заметить, что это в основном начало рабочего дня, обеденный перерыв и конец рабочего дня (все едут на работу, с работы, на обед)__ 

```python
df['Время'].value_counts()[:10]
```

```python
df = df.astype({'Время' : 'datetime64' })
sns.set(rc={'figure.figsize':(22, 10)})
sns.lineplot(data=df['Время'].value_counts())
```

__Глянем теперь на корреляцию между числовыми признаками:__

```python
sns.pairplot(df)
```
